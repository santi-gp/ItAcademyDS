# Solución

1. Aplicar test de normalidad a las variables numericas.

Para determinar si los datos siguen una distribución normal utilizaremos la prueba de Anderson-Darling. Además, se van a omitir las variables CHAS y RAD (categóricas).

````python
columns=list(set(list(df.columns))-set({'CHAS','RAD'}))
````

##### Hipótesis nula
H0: los datos de las variables siguen una distribución normal

````python
from scipy.stats import anderson

# normality test
for i in columns:
    result = anderson(df[i],dist='norm')
    print([i])
    print('Statistic: %.3f' % result.statistic)
    p = 0
    for i in range(len(result.critical_values)):
        sl, cv = result.significance_level[i], result.critical_values[i]
        if result.statistic < result.critical_values[i]:
            print('%.3f: %.3f, los datos parecen normales (no se rechaza H0)' % (sl, cv))
        else:
            print('%.3f: %.3f, los datos no parecen normales (rechazar H0)' % (sl, cv))
````

Valores más grandes del estadístico de Anderson-Darling indican que los datos no siguen la distribución normal.
De esta prueba se puede afirmas que nuestras variables no siguen una distribución normal.

### Transformación de variables

#### Normalización MinMax

El concepto de normalización trata del proceso de ajustar los valores medidos en diferentes escalas a escalas comunes antes de realizar estadísticas descriptivas, y se denota mediante la siguiente ecuación: 

$$
x_{nor}=\frac{x-x_{min}}{x_{max}-x_{min}}
$$

#### Normalización estándar

La normalización estándar es similar a la normalización excepto que tiene una media de 0 y una desviación estándar de 1. La normalización estándar se denota mediante la siguiente ecuación: 

$$
x_{st}=\frac{x-\mu}{\sigma}
$$

Por Ejemplo, para  crear dos nuevas columna una normalizada y otra estandarizada de la columna **TAX**, podemos utilizar el siguiente código: 

```python
df['TAX_NOR']=(df.TAX-df.TAX.min())/(df.TAX.max()-df.TAX.min())
df.head()
```

```python
df['TAX_SN']=(df.TAX-df.TAX.mean())/(df.TAX.std())
df.head()
```

Sin embargo, haremos uso de la libreria **sklearm** de python para obtener estas nuevas columnas de la siguiente manera:

#### 3 Escalar columnas con el método MinMaX
Vamosa transformar aquellas columnas con presentan valore atípicos. Para ello primero haremos una copia de nuestro data frame de la siguiente manera:

````python
df_trans=df.copy()
````

Se puede ver en los **boxplots** del gráfico anterior que las variables TAX, NOX, INDUS y AGE no tienem valores atipicos. Por lo tanto, normalizaremos estas columnas con el método **MinMaxScaler**().

````python
normalize=MinMaxScaler()
df_trans[['TAX','INDUS','AGE']]=normalize.fit_transform(df_trans[['TAX','INDUS','AGE']])
df_trans.head()
````

#### 4. Escalar columnas con el método RobustScaler

Se vio que las variables RM, CRIM, B, LSTAT, MED, DIS, ZN y PTRATIO tienen valores atípicos. Sin embargo, en el caso de variables con valores atípicos podemos utilizar la transformación **RobustScaler()** de la librería **Sklearn**.

````python
list_outliers=['RM','CRIM','B','LSTAT','MEDV','DIS','ZN','PTRATIO']
robust=RobustScaler()
df_trans[list_outliers]=robust.fit_transform(df_trans[list_outliers])
df_trans.head()
````

#### 5 Crear variables ficticias (dummys)
Vamos a crear varaibles ficticias de la variable discreta RAD, primero vamos a imprimir sus valores únicos

````python
df_trans.RAD.unique()
````

````python
# Creamos las variables
df_trans = pd.get_dummies(data=df_trans, columns=['RAD'])
df_trans.head()
````

Podemos  ver en el nuevo data frame que se han creado nueve columnas. 

Básicamente lo que hace la función es crear una columna por cada valor diferente de cada celda, separándolos por el caracter que nosotros especifiquemos (para este caso RAD_num), y rellenar dicha columna con ceros y unos. 0 si no está el valor presente en una celda y 1 si está presente.
